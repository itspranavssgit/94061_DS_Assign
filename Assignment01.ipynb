## 2. Write a NumPy program to create an element-wise comparison (greater,  greater_equal, less and less_equal) of two given arrays.


```python
a = np.array([3, 7, 1, 9])
b = np.array([2, 7, 5, 4])

print("a > b :", np.greater(a, b))
print("a >= b:", np.greater_equal(a, b))
print("a < b :", np.less(a, b))
print("a <= b:", np.less_equal(a, b))
```

## 6. Write a NumPy program to get the element-wise remainder of an array of numbers from 20 to 50 which are divisible by 3.


```python
arr = np.arange(20, 51)
div_by_3 = arr[arr % 3 == 0]
remainder = np.remainder(div_by_3, 3)

print("Numbers divisible by 3:", div_by_3)
print("Remainder when divided by 3:", remainder)
```

## 7. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
   Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]


```python
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 9])
print("Addition:")
print(s1 + s2)

print("\nSubtraction:")
print(s1 - s2)

print("\nMultiplication:")
print(s1 * s2)

print("\nDivision:")
print(s1 / s2)
```

## 10. i) Create a list of empids and names (10 employees).
    ii) Convert list into Series.
    iii) Print type of Series.
    iv) Convert Series into DataFrame.
    v) a) Display all records.
       b) Display first five records.
       c) Display last five records.
       d) Display first two and last two records.
       e) Display name of 5th employee.
       f) Display record of 4th employee.
    vi) Save the record in csv file (MyRecord.csv)



```python
# i) Create a list of empids and names (10 employees)
emp_ids = [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
emp_names = ['Amit', 'Ravi', 'Neha', 'Pooja', 'Rahul',
             'Sneha', 'Anil', 'Kiran', 'Priya', 'Vikas']

emp_list = list(zip(emp_ids, emp_names))

# ii) Convert list into Series
emp_series = pd.Series(emp_list)

# iii) Print type of Series
print("Type of emp_series:", type(emp_series))

# iv) Convert Series into DataFrame
df = pd.DataFrame(emp_series, columns=['Employee'])

# Split tuple into separate columns
df[['EmpID', 'Name']] = pd.DataFrame(df['Employee'].tolist(), index=df.index)
df.drop(columns='Employee', inplace=True)

# v) a) Display all records
print("\nAll Records:")
print(df)

# v) b) Display first five records
print("\nFirst Five Records:")
print(df.head())

# v) c) Display last five records
print("\nLast Five Records:")
print(df.tail())

# v) d) Display first two and last two records
print("\nFirst Two Records:")
print(df.head(2))

print("\nLast Two Records:")
print(df.tail(2))

# v) e) Display name of 5th employee
print("\nName of 5th Employee:")
print(df.loc[4, 'Name'])

# v) f) Display record of 4th employee
print("\nRecord of 4th Employee:")
print(df.loc[3])

# vi) Save the record in csv file
df.to_csv("MyRecord.csv", index=False)

print("\nData saved successfully to MyRecord.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load mtcars dataset
df = pd.read_csv(r"C:\Users\Anant\Desktop\Class Work\mtcars.csv")

# i) Line Plot
plt.figure()
plt.plot(df['mpg'])
plt.title("Line Plot of MPG")
plt.xlabel("Index")
plt.ylabel("MPG")
plt.show()

# ii) Bar Graph
plt.figure()
plt.bar(df.index, df['mpg'])
plt.title("Bar Graph of MPG")
plt.xlabel("Car Index")
plt.ylabel("MPG")
plt.show()

# iii) Pie Chart
cyl_counts = df['cyl'].value_counts()
plt.figure()
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart of Cylinders")
plt.show()

# iv) Boxplot
plt.figure()
plt.boxplot(df['mpg'])
plt.title("Boxplot of MPG")
plt.ylabel("MPG")
plt.show()

# v) Pairplot
sns.pairplot(df[['mpg', 'disp', 'hp', 'wt']])
plt.show()

# vi) Scatter Plot
plt.figure()
plt.scatter(df['wt'], df['mpg'])
plt.title("Scatter Plot of Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()

# vii) Histogram
plt.figure()
plt.hist(df['mpg'], bins=10)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```
